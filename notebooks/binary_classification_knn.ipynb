```python
from sklearn import preprocessing
from sklearn.neighbors import KNeighborsClassifier
from tqdm import tqdm
import cv2
import numpy as np
import pandas as pd
import csv
```

# Data Preprocessing 

```python
dataset = pd.read_csv(icml_face_data_path)
dataset[" Usage"].unique()
```

```python
train_dataset = dataset[dataset[" Usage"] =="Training"]
test_dataset = dataset[dataset[" Usage"]=="PublicTest"]
test_dataset.head()
```

```python
def convert_to_image(string_data):
  image_numpy = []
  for i in tqdm(range(len(string_data))):
    image_numpy.append(np.array(string_data[i].split()).astype(np.int))
  return np.array(image_numpy)
```

```python
def to_binary_classification(labels):
    binary_label = []
    for i in tqdm(range(len(labels))):
        binary_label.append(1 if labels[i] == 0 else 0)
    return binary_label
```

```python
train_features = convert_to_image(train_dataset[' pixels'].to_numpy())
train_labels = to_binary_classification(train_dataset['emotion'])
```

```python
test_features = convert_to_image(test_dataset[' pixels'].to_numpy())
test_labels = to_binary_classification(list(test_dataset['emotion']))
```

# KNN Model

```python
knn_model = KNeighborsClassifier()
knn_model.fit(train_features, train_labels)
```

```python
knn_model.score(test_features, test_labels)
```
